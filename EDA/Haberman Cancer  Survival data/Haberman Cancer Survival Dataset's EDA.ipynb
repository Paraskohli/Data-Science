```python
data_survived=data.loc[data['survived_status']=='survived']
data_dead=data.loc[data['survived_status']=='dead']
```

```python
#pdf, cdf of survived people
counts,bin_edges=np.histogram(data_survived['axil_nodes'],bins=10,density=True)
pdf=counts/sum(counts)
print(pdf)
print(bin_edges)

#compute cdf
cdf=np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)

plt.xlabel('axil_nodes')
plt.show();
```

```python
#pdf, cdf of dead people

counts,bin_edges=np.histogram(data_dead['axil_nodes'],bins=10,density=True)
pdf=counts/sum(counts)
print(pdf)
print(bin_edges)

#compute cdf
cdf=np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)

plt.xlabel('axil_nodes')
plt.show();
```

```python
# Quantiles
print(np.percentile(data_survived['axil_nodes'],np.arange(0,100,25)))
print(np.percentile(data_dead['axil_nodes'],np.arange(0,100,25)))
```

From here we ensure that,having <b>no. of axil nodes = 0 </b>,approx 75 percent people 'survive' and approximatly 25 percent of people 'dead'.
